# Análise Exploratória


```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Pelo Google Colab
#from google.colab import files
#arq = files.upload()
```

```python
#Criando o nosso Dataframe
df = pd.read_excel("../datasets/AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando o tipo de dados
df.dtypes
```

```python
#Qual foi a receita total ?
df['Valor Venda'].sum()
```

```python
#Qual o custo total ?
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head(1)
```

```python
#Qual o Custo total?
round(df['Custo'].sum(),2)
```

```python
df['lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head(1)
```

```python
#Total Lucro
round(df['lucro'].sum(),2)
```

```python
#Criando Coluna com tempo de envio
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

** Queremos Saber a média do tempo de envio para cada Marca e para isso precisamos transformar a coluna tempo envio  em númerica **

```python
#Extraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] -df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna tempo_envio
df['Tempo_envio'].dtype
```

```python
#Média do tempo de envio por Marca
df.groupby(df['Marca'])['Tempo_envio'].mean()
```

**Missing Values**

```python
#vericando se temos valores de dados faltantes
df.isnull().sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Vamos agrupar por ano e por marca
df.groupby([df['Data Venda'].dt.year,"Marca"])['lucro'].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
